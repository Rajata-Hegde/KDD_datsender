```python
def attack_class(label):
    if label == 'normal':
        return 'Normal'
    elif label in ['back','land','neptune','pod','smurf','teardrop','mailbomb',
                   'apache2','processtable','udpstorm']:
        return 'DoS'
    elif label in ['satan','ipsweep','nmap','portsweep','mscan','saint']:
        return 'Probe'
    elif label in ['guess_passwd','ftp_write','imap','phf','multihop',
                   'warezmaster','warezclient','spy','xlock','xsnoop',
                   'snmpguess','snmpgetattack','httptunnel','sendmail','named']:
        return 'R2L'
    else:
        return 'U2R'

df['attack_class'] = df['label'].apply(attack_class)
```

```python
from sklearn.preprocessing import LabelEncoder

le_target = LabelEncoder()
df['attack_class_encoded'] = le_target.fit_transform(df['attack_class'])

print(dict(zip(le_target.classes_, range(len(le_target.classes_)))))
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['label', 'attack_class', 'attack_class_encoded'], axis=1)
y = df['attack_class_encoded']

X_train_base, X_test, y_train_base, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
df_gan_train = df.loc[X_train_base.index].copy()
```

```python
# Find minority classes
minority_classes = (
    df_gan_train['attack_class_encoded']
    .value_counts()
    .nsmallest(2)
    .index
)

print("Minority class codes:", minority_classes.tolist())

minority_df = df_gan_train[
    df_gan_train['attack_class_encoded'].isin(minority_classes)
]

print(minority_df['attack_class_encoded'].value_counts())
```

```python
# -----------------------------------------
# REMOVE STRING COLUMNS BEFORE CTGAN
# -----------------------------------------

cols_to_drop = ['label', 'attack_class']

for col in cols_to_drop:
    if col in minority_df.columns:
        minority_df = minority_df.drop(columns=[col])

print("Remaining dtypes in minority_df:")
print(minority_df.dtypes)
```

```python
from ctgan import CTGAN

categorical_cols = ['protocol_type', 'service', 'flag']
discrete_columns = categorical_cols + ['attack_class_encoded']

# Drop difficulty if present
if 'difficulty' in minority_df.columns:
    minority_df = minority_df.drop(columns=['difficulty'])

ctgan = CTGAN(
    epochs=30,
    batch_size=500,
    verbose=True
)

ctgan.fit(minority_df, discrete_columns)
```

```python
class_counts = df_gan_train['attack_class_encoded'].value_counts()
majority_count = class_counts.max()

synthetic_samples = []

for cls in minority_classes:
    current = class_counts[cls]
    needed = majority_count - current

    print(f"Generating {needed} samples for class {cls}")

    batches = []
    batch_size = 1000

    while needed > 0:
        n = min(batch_size, needed)
        sample = ctgan.sample(
            n,
            condition_column='attack_class_encoded',
            condition_value=cls
        )
        batches.append(sample)
        needed -= n

    synthetic_samples.append(pd.concat(batches, ignore_index=True))
```

```python
df_synthetic = pd.concat(synthetic_samples, ignore_index=True)

df_gan_balanced = pd.concat(
    [df_gan_train, df_synthetic],
    ignore_index=True
)

print(df_gan_balanced['attack_class_encoded'].value_counts())
```

```python
print("=" * 70)
print("CHECKING GAN RESULTS")
print("=" * 70)

# Save the GAN dataset to a CSV file
df_gan_balanced.to_csv('balanced_dataset_gan.csv', index=False)

# Load the GAN dataset
df_gan_check = pd.read_csv('balanced_dataset_gan.csv')

# Check class distribution
gan_counts = df_gan_check['attack_class_encoded'].value_counts().sort_index()
print("\nGAN Dataset Class Distribution:")
for cls, count in gan_counts.items():
    label = le_target.inverse_transform([cls])[0]
    print(f"  {label}: {count:,} samples")

# Check if it's balanced (all classes should have similar counts)
if gan_counts.std() / gan_counts.mean() < 0.1:
    print("✅ GAN dataset is properly balanced!")
else:
    print("⚠️ GAN dataset is NOT properly balanced")
    print("   Standard deviation too high - some classes have very different counts")
```

```python
X_train_gan = df_gan_balanced.drop(
    ['label', 'attack_class', 'attack_class_encoded'],
    axis=1
)

y_train_gan = df_gan_balanced['attack_class_encoded']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_gan_scaled = scaler.fit_transform(X_train_gan)
X_test_scaled = scaler.transform(X_test)
```

```python
# ============================================
# SIMPLE NN BASELINE (FOR SMOTE vs GAN)
# ============================================

import torch
import torch.nn as nn
import torch.optim as optim

# Ensure reproducibility
torch.manual_seed(42)

# -----------------------------
# Simple Neural Network
# -----------------------------
class SimpleNN(nn.Module):
    def __init__(self, input_size, num_classes):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, num_classes)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x

# -----------------------------
# Training & Evaluation Function
# -----------------------------
def train_simple_model(X_train, X_test, y_train, y_test, name, epochs=20):
    print(f"\n==============================")
    print(f"Training {name} model")
    print(f"==============================")

    # Convert to tensors
    X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.long)
    X_test_tensor  = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor  = torch.tensor(y_test, dtype=torch.long)

    # Initialize model
    model = SimpleNN(
        input_size=X_train.shape[1],
        num_classes=len(torch.unique(y_train_tensor))
    )

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # -----------------------------
    # Training loop
    # -----------------------------
    for epoch in range(epochs):
        model.train()

        optimizer.zero_grad()
        outputs = model(X_train_tensor)
        loss = criterion(outputs, y_train_tensor)
        loss.backward()
        optimizer.step()

        if (epoch + 1) % 5 == 0:
            print(f"Epoch [{epoch+1}/{epochs}] - Loss: {loss.item():.4f}")

    # -----------------------------
    # Evaluation
    # -----------------------------
    model.eval()
    with torch.no_grad():
        outputs = model(X_test_tensor)
        _, predictions = torch.max(outputs, 1)
        accuracy = (predictions == y_test_tensor).float().mean().item()

    print(f"✅ {name} Test Accuracy: {accuracy:.4f}")
    return accuracy
```

```python
# ----------------------------- # Training with SMOTE data # -----------------------------
smote_accuracy = train_simple_model(
    X_train_sm, X_test_scaled, le_target.transform(y_train_sm),
    y_test.values,
    name="SMOTE Baseline NN"
)

# ----------------------------- # Training with GAN data # -----------------------------
gan_accuracy = train_simple_model(
    X_train_gan_scaled, X_test_scaled, y_train_gan,
    y_test.values,
    name="GAN Baseline NN"
)

print("\n--- Summary ---")
print(f"SMOTE Model Test Accuracy: {smote_accuracy:.4f}")
print(f"GAN Model Test Accuracy:   {gan_accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a DataFrame for plotting
accuracy_data = {
    'Model': ['SMOTE Baseline NN', 'GAN Baseline NN'],
    'Accuracy': [smote_accuracy, gan_accuracy]
}
accuracy_df = pd.DataFrame(accuracy_data)

plt.figure(figsize=(8, 5))
sns.barplot(x='Model', y='Accuracy', data=accuracy_df, palette='viridis')
plt.ylim(0, 1) # Set y-axis limit from 0 to 1 for accuracy
plt.title('Comparison of SMOTE and GAN Model Accuracies')
plt.ylabel('Test Accuracy')
plt.xlabel('Data Augmentation Method')

# Add accuracy values on top of the bars
for index, row in accuracy_df.iterrows():
    plt.text(row.name, row.Accuracy + 0.01, f'{row.Accuracy:.4f}', color='black', ha='center')

plt.show()
```
